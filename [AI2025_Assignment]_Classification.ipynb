# Assignments

The assignment of using the chemical composition data of wine to freely train your own classification model among several classifications to identify the type of wine and to check the accuracy of the model.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

##Data load and Preprocessing

```python
from sklearn.datasets import load_wine

wine = load_wine()
```

##Inspecting Data and Splitting Train_set, Test_set

* test_size = 0.3
* Check how the data is organized through print(wine['DESCR'])

```python
# Check the Dataset Descriptions
print(wine['DESCR'])
```

```python
print(wine.feature_names)
print(wine.target_names)
```

### Please define X and Y to predict `class` through `wine_df`.

```python
wine_df = pd.DataFrame(wine['data'], columns=wine.feature_names)
wine_df['class'] = wine.target

# Setting X, Y
X = wine_df.drop('class', axis=1)
Y = wine_df['class']
```

```python
print(X.head())
```

```python
print(Y.head())
print(Y.tail())
```

```python
wine_df.info()
wine_df['proline_log'] = np.log1p(wine_df['proline'])
wine_df = wine_df.drop(columns=['proline'])
wine_df.describe()
```

### Split train and test data through `train_test_split`.
- test_size = 0.3, random_state = 0

```python
from sklearn.model_selection import train_test_split

# Splitting Train/Test data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

##Build Models Using Various Models

```python
from sklearn.ensemble import RandomForestClassifier

# Generate Model
model = RandomForestClassifier(
    n_estimators=200,
    criterion='gini',
    max_depth=None,
    min_samples_split=2,
    min_samples_leaf=1,
    max_features='sqrt',
    bootstrap=True,
    random_state=0,
    n_jobs=-1
)

# Fitting using scaled training data
model.fit(X_train_scaled, Y_train)

# Predict
Y_predict = model.predict(X_test_scaled)
```

##Check the fitted model's performance


```python
from sklearn.metrics import classification_report
print(classification_report(Y_test, Y_predict))
```

### Print the Accuracy.

```python
# model accuarcy
accuracy = model.score(X_test_scaled, Y_test)
print("Accuracy: {0:.3f}".format(accuracy))
```

## 하이퍼파라미터 튜닝 (GridSearchCV)

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import StratifiedKFold, cross_val_score, ParameterGrid
from sklearn.metrics import accuracy_score
import numpy as np

rf = RandomForestClassifier(random_state=0, n_jobs=-1)
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=0)

param_grid_stage1 = {
    "n_estimators": [50, 100, 200],
    "max_depth": [4, 8, 12, None],
    "min_samples_split": [2, 5, 10],
    "min_samples_leaf": [1, 2, 4],
    "max_features": ["sqrt", "log2", None]
}

best_score = -np.inf
best_params = None
grid1 = list(ParameterGrid(param_grid_stage1))
for i, p in enumerate(grid1, 1):
    rf.set_params(**p)
    scores = cross_val_score(rf, X_train_scaled, Y_train, cv=cv, scoring="accuracy", n_jobs=-1)
    acc = scores.mean()
    if acc > best_score:
        best_score, best_params = acc, p
    print(f"[Stage1 {i}/{len(grid1)}] acc={acc:.3f} best={best_score:.3f}")

center_n = best_params["n_estimators"]
center_depth = best_params["max_depth"] if best_params["max_depth"] is not None else 12
center_split = best_params["min_samples_split"]
center_leaf = best_params["min_samples_leaf"]

def rng(center, left, right, low, high, allow_none=False):
    vals = sorted(set([v for v in range(max(low, center-left), min(high, center+right)+1)]))
    if allow_none and None not in vals:
        vals = vals + [None]
    return vals

n_upper = min(500, center_n + 200)
stage2_grid = {
    "n_estimators": rng(center_n, 50, 100, 10, n_upper),
    "max_depth": [None] + rng(center_depth if center_depth is not None else 12, 2, 4, 2, 30),
    "min_samples_split": rng(center_split, 2, 5, 2, X_train_scaled.shape[0]//2),
    "min_samples_leaf": rng(center_leaf, 1, 3, 1, max(5, X_train_scaled.shape[0]//10)),
    "max_features": [best_params["max_features"]]
}

best_score2 = best_score
best_params2 = best_params
grid2 = list(ParameterGrid(stage2_grid))
for i, p in enumerate(grid2, 1):
    rf.set_params(**p)
    scores = cross_val_score(rf, X_train_scaled, Y_train, cv=cv, scoring="accuracy", n_jobs=-1)
    acc = scores.mean()
    if acc > best_score2:
        best_score2, best_params2 = acc, p
    print(f"[Stage2 {i}/{len(grid2)}] acc={acc:.3f} best={best_score2:.3f}")

rf.set_params(**best_params2)
rf.fit(X_train_scaled, Y_train)
Y_pred = rf.predict(X_test_scaled)
test_acc = accuracy_score(Y_test, Y_pred)

print("\n결과 요약")
print("Stage1 best params:", best_params)
print(f"Stage1 best CV acc: {best_score:.3f}")
print("Stage2 best params:", best_params2)
print(f"Stage2 best CV acc: {best_score2:.3f}")
print(f"Test acc: {test_acc:.3f}")
```
